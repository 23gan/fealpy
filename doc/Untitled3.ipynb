```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x  = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)[:, np.newaxis]
z = np.sin(x)**10 + np.cos(10 + y*x)*np.cos(x)
```

```python
plt.imshow(z, origin='lower', extent=[0, 5, 0, 5], cmap='viridis')
plt.colorbar()
```
